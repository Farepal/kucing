```python
fertility = 'Fertility rate - Sex: all - Age: all - Variant: estimates'
gdp = 'GDP per capita (output, multiple price benchmarks)'
population = 'Population (historical estimates)'
```

```python
df_2019_clean = df_2019.dropna(subset=[fertility, gdp])
df_2019_clean
# cara output semuanya
# cara jadi excel gimana ya?
```

```python
# linear regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = df_2019_clean[gdp].values.reshape(-1, 1)
y = df_2019_clean[fertility].values.reshape(-1, 1)
X = np.log(X)
model.fit(X, y)


print("Coefficient: ", model.coef_)
print("Intercept: ", model.intercept_)
print("Correlation Coefficient: ", np.corrcoef(X[:,0], y[:,0])[0,1])

# plot
plt.figure(figsize=(10, 10))
plt.scatter(X, y, color='black')
plt.plot(X, model.predict(X), color='blue', linewidth=3)
plt.xlabel('Natural logarithm of GDP per capita (2019)')
plt.ylabel('Children per Woman (2019)')
plt.title('Fertility rate vs Natural logarithm of GDP per capita')
plt.show()

b = model.coef_[0][0]
a = model.intercept_[0]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
X = df_2019_clean[gdp].values.reshape(-1, 1)
y = df_2019_clean[fertility].values.reshape(-1, 1)

plt.scatter(X, y, color='black')
g = np.linspace(1000, 100000, 10000)  # Menetapkan rentang nilai x

y = b * np.log(g) + a  # Persamaan fungsi
# Plot data
plt.plot(g, y, label='ln(x)')

# Menambahkan label dan judul
plt.xlabel('GDP per Capita')
plt.ylabel('Children per Woman')
plt.title('Children per Woman vs GDP per Capita')

# Menampilkan legenda
plt.legend()

# Menampilkan grafik
plt.show()
```
