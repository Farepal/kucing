```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft

# Parameter gerak melingkar
omega_0 = 2*np.pi  # Kecepatan sudut konstan

# Waktu
t = np.linspace(0, 2*np.pi, 1000)  # Satu periode gerak melingkar

# Sinyal sudut
theta_t = omega_0 * t

# Transformasi Fourier
theta_f = fft(theta_t)

# Plot sinyal sudut
plt.subplot(2, 1, 1)
plt.plot(t, theta_t)
plt.title('Sinyal Sudut Gerak Melingkar')
plt.xlabel('Waktu')
plt.ylabel('Sudut')

# Plot spektrum frekuensi
plt.subplot(2, 1, 2)
freq = np.fft.fftfreq(len(t), t[1] - t[0])
plt.plot(freq, np.abs(theta_f))
plt.title('Spektrum Frekuensi')
plt.xlabel('Frekuensi')
plt.ylabel('Amplitudo')

plt.tight_layout()
plt.show()
```
